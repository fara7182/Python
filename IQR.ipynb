```python
q1,q3,q4 = np.percentile(df['Время_выполнения'],(25,75,100))
print(q1,q3,q4)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
lower_hence = q1 - (iqr*1.5)
higher_hence = q3 + (iqr*1.5)
print(lower_hence, higher_hence)
```

```python
x = df['Время_выполнения']
```

```python
c = 0
for dn in x:
    if (dn > higher_hence):
        c = c+1
```
